```python
from ColabTurtle.Turtle import *

initializeTurtle(initial_speed=5)

angulos = [160, -43, 270, -97, -43, 200, -940, 17, -86]
direcao_final = 0

for angulo in angulos:
    left(angulo)
    forward(100)
    direcao_final += angulo

direcao_final = direcao_final % 360
print("A direção final do pirata é:", direcao_final, "graus")
```
